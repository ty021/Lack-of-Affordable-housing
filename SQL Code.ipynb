# LACK OF AFFORDABLE HOUSING 1

Which towns in the dataset 1 have highest percentage of affordable housing unit (Percent Affordable)

```python
SELECT
   Town,
  MAX ("Percent Affordable") AS More_Affordable
FROM
  `adta5240group1-443622.Lack_of_ah1.dataset 1`
GROUP BY
  Town
ORDER BY
  More_Affordable DESC
LIMIT 10;
```

Out of all town in the dataset 1 which town has the highest number of the government-assisted units for each year

```python
SELECT
  Year,
  Town,
  MAX ('Government Assisted') AS More_Gov_Assist
FROM
  `adta5240group1-443622.Lack_of_ah1.dataset 1`
GROUP BY
  Year, Town
ORDER BY
  Year ASC, More_Gov_Assist DESC;
```

# LACK OF AFFORDALE HOUSING 2

From the dataset 2 the highest total loan amount is present for which cities for all housing projects

```python
SELECT
  City,
  SUM (CAST (Loan_Amount AS NUMERIC)) AS tot_loan_amt
FROM
  `adta5240group1-443622.Lack_of_ah2.dataset_2`
GROUP BY
  City
ORDER BY
  tot_loan_amt DESC
LIMIT 50;
```

From the dataset 2 based on the construction type what is the average development cost

```python
SELECT
  Construction_Type,
  AVG(Total_Development_Cost) AS avg_cost_of_develop 
FROM
  `adta5240group1-443622.Lack_of_ah2.dataset_2`
GROUP BY
  Construction_Type
ORDER BY
  avg_cost_of_develop DESC;
```

# LACK OF AFFORDABLE HOUSING 3

In the dataset 3 what are the highest number of assisted units give top 10 (Project Number of Assisted Units)

```python
SELECT
  `Property Name` AS Prop_name,
  `Project City Address` AS Proj_city,
  `Project State` AS Proj_state,
  `Project Number of Assisted Units` AS Assist_units
FROM
  `adta5240group1-443622.Lack_of_ah3.dataset_3`
ORDER BY
  Assist_units DESC
LIMIT 10;
```

In the dataset 3 what are the states that have the highest total number of assisted units

```python
SELECT
  `Project State` AS Proj_state,
  SUM (`Project Number of Assisted Units`) AS tot_assist_unit
FROM
  `adta5240group1-443622.Lack_of_ah3.dataset_3`
GROUP BY
  Proj_state
ORDER BY
  tot_assist_unit DESC
LIMIT 10;
```
